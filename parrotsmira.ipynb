```python
def read_fasta_file(fasta_file):              #задать функцию
    fasta = {}                                #создать пустой словарь fasta
    header = None                             #создать пустую переменную header с неопределенным значением
    with open(fasta_file) as fh:              #открыть файл
        for line in fh:                       #для каждого объекта(тут объект=строка) в файле
            line = line.strip()               #удалить из строки пробелы/перенос строки
            if line.startswith(">"):          #если строка начитается с >
                if header:                    #если значение header не является пустым
                    fasta[header] = "".join(seq) #записать в словарь fasta header в качестве ключа и seq в качестве значения
                header = line[1:]             #переписать значение header на значение текущего объекта line
                seq = []                      #создать пустой список seq
            else:
                seq.append(line)              #добавить в список seq значение line
        if header:                            #если значение header не является пустым
            fasta[header] = "".join(seq)      #записать в словарь fasta header в качестве ключа и seq в качестве значения
    return fasta                              #конец тела функции
```

```python
fasta = read_fasta_file(fasta_file)
```
